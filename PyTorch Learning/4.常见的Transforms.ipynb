```python
from PIL import Image 
from torchvision import transforms
from torch.utils.tensorboard import SummaryWriter

img = Image.open("../dataset/练手数据集/train/ants_image/0013035.jpg")
print(img)
```

```python
class Person:
    def __call__(self,name):
        print("__init__"+name)
    def hello(self,name):
        print("hello"+name)
        
person = Person()
person("zhangsan")
person.hello("lisi")
```

```python
writer = SummaryWriter("logs")

trans_tensor = transforms.ToTensor()
img_tensor = trans_tensor(img)
writer.add_image("ToTensor",img_tensor)


# Normalize
print(img_tensor[0][0][0])
trans_norm = transforms.Normalize([6,3,2],[9,3,5])
img_norm = trans_norm(img_tensor)
print(img_norm[0][0])
writer.add_image("Normalize",img_norm,2)
```

```python
# Resize
print(img.size)
trans_size = transforms.Resize((512,512))
img_resize = trans_size(img)
img_resize = trans_tensor(img_resize)
writer.add_image("Resize",img_resize,0)
print(img_resize)
```

```python
# Compose - Resize
# 会改变长和宽的比例
trans_resize_2 = transforms.Resize(512)
trans_compose = transforms.Compose([trans_resize_2,trans_tensor])
img_resize_2 = trans_compose(img)
writer.add_image("Resize",img_resize_2,1)

```

```python
# RandomCrop
# 随机裁剪

trans_random = transforms.RandomCrop((500,600))
trans_compose_2 = transforms.Compose([trans_random,trans_tensor])
for i in range(10):
    img_crop = trans_compose_2(img)
    writer.add_image("RandomCrop",img_crop,i)
```
